## Word Count
Checking to see if any language has a large difference in the number of words spoken

```python
df["Word Count"] = [len(x.split()) for x in df['text']]
#print((df["Word Count"].describe()))
df.boxplot(by = ['accent'], column = ['Word Count'])
```

## Beginning to deal with large unbalance in the dataset
We will combine a lot of the accents with lower numbers of samples into a single 'other' category

```python
df.accent.replace(['singapore','southatlandtic',
                   'bermuda','malaysia',
                   'hongkong','wales',
                   'philippines','newzealand',
                   'ireland','african',
                   'scotland'],
                   'other',
                   inplace = True)

df.dropna(subset = ['accent'],inplace= True)#drop rows with no accent information

plt.bar(df['accent'].unique()[:],df.accent.value_counts())

plt.show()
```

## Dealing with missing/non-numeric data

```python
df.drop(columns=['up_votes','down_votes','text','duration','Word Count'],inplace  = True) # drop uncorrelated columns
df.age.replace(['teens','twenties',     #Convert age to numbers and fill NA's
                'thirties','fourties',
                'fifties','sixties',
                'seventies','eighties'],
               [1,2,3,4,5,6,7,8],
               inplace = True
               )
df.age.fillna(0,inplace = True)

df.gender.replace(['male','female'],[0,1],inplace = True) #convert gender to 0 or 1

df.gender.replace(['other'],
               [-1],
               inplace = True)#Get rid of all other unwanted data types in the gender column

df.gender.fillna(-1,inplace = True)


le = preprocessing.LabelEncoder()
df['accent'] = le.fit_transform(df.accent)
df.info()
```

## Using a KDE plot to understand initial dataset feature significance

```python
sns.set_theme()
plt.subplots(1,2,figsize=(20,10))
for k in range(1,3):
    plt.subplot(1,2,k)
    plt.title(df.columns[k])
    for lab in df.accent.unique():
        sns.kdeplot(df.loc[df['accent'] == lab, df.columns[k]], label=lab)
plt.legend(le.inverse_transform(df.accent.unique()))
plt.figure(dpi=300)
plt.show()
```

## Conclusion
The dataset contains no features in its current state that will indicate accent. Research will need to be done about developing features directly from the .wav files. Another issue will be the imbalance of the data during training.
